2. Una tienda ofrece un descuento del 15% sobre el total
de la compra y un cliente desea saber cuánto deberá
pagar finalmente por su compra.

```python
# muestra el valor de la compra sin descuento
compra = float(input("digite el valor de la compra:"))
# muestra el descuento que se aplicara a la compra
descuen = compra * 0.15
totalpag = compra - descuen
print("el descuento de la compra es de = ${:,.0f} y el valor a pagar es de = ${:,.0f}".format(descuen, totalpag))
```

3. Un alumno desea saber cuál será su calificación final en
la materia de Algoritmos. Dicha calificación se compone
de tres exámenes parciales.

```python
# muestra calificacion de examenes
calif1 = float(input("Ingresa calificacion1:"))
calif2 = float(input("Ingresa calificacion2:"))
calif3 = float(input("Ingresa calificacion3:"))
# muestra la calificacio final 
califin = round((calif1 + calif2 + calif3) /3, 2)
print(f"calificacion final materia algoritmos ={califin} ")
```
